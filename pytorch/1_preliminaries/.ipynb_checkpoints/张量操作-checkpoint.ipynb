# 数据操作


首先，我们介绍$n$维数组，也称为*张量*（tensor）。

无论使用哪个深度学习框架，它的*张量类*（在MXNet中为`ndarray`，
在PyTorch和TensorFlow中为`Tensor`）都与Numpy的`ndarray`类似。
但深度学习框架又比Numpy的`ndarray`多一些重要功能：
首先，GPU很好地支持加速计算，而NumPy仅支持CPU计算；
其次，张量类支持自动微分。
这些功能使得张量类更适合深度学习。
如果没有特殊说明，本书中所说的张量均指的是张量类的实例。


```python
# 虽然它被称为PyTorch，但是代码中使用`torch`而不是`pytorch`。
import torch
```

## 创建张量

[**张量表示由一个数值组成的数组，这个数组可能有多个维度**]。
具有一个轴的张量对应数学上的*向量*（vector）；
具有两个轴的张量对应数学上的*矩阵*（matrix）；
具有两个轴以上的张量没有特殊的数学名称。


首先，这个行向量包含以0开始的前12个整数，它们默认创建为整数。也可指定创建类型为浮点数。张量中的每个值都称为张量的 *元素*（element）。例如，张量 `x` 中有 12 个元素。除非额外指定，新的张量将存储在内存中，并采用基于CPU的计算。


```python
# `arange` 创建一个行向量 `x`
X = torch.arange(12)
X
```

```python
# 全0
a = torch.zeros((2, 3, 4))
b = torch.zeros_like(X)
# 全1
c = torch.ones((2, 3, 4))
d = torch.ones_like(X)

a,b,c,d
```

```python
# 其中的每个元素都从均值为0、标准差为1的标准高斯分布（正态分布）中随机采样。
torch.randn(3, 4)
```

```python
# Python列表
torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
```

## 张量的属性

```python
# 张量（沿每个轴的长度）的形状
X.shape
```

```python
# 张量中元素的总数，即形状的所有元素乘积
X.numel()
```

```python
# 要想改变一个张量的形状
torch.arange(12).reshape(3, 4)
```

我们不需要通过手动指定每个维度来改变形状。
也就是说，如果我们的目标形状是（高度,宽度），
那么在知道宽度后，高度会被自动计算得出，不必我们自己做除法。
在上面的例子中，为了获得一个3行的矩阵，我们手动指定了它有3行和4列。
幸运的是，我们可以通过`-1`来调用此自动计算出维度的功能。
即我们可以用`x.reshape(-1,4)`或`x.reshape(3,-1)`来取代`x.reshape(3,4)`。


在这里，最外层的列表对应于轴0，内层的列表对应于轴1。


## 按元素（elementwise）运算



```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])

a = x + y
b = x - y,
c = x / y
d = x * y   # 按元素相乘，不是内积
e = x ** y  # 求幂运算

a, b, c, d, e
```

```python
torch.exp(x)
```

```python
# 连接张量
X = torch.arange(12).reshape((3, 4))
Y = torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])

# 按外层的列
a = torch.cat((X, Y), dim=0)
# 按内层的列
b = torch.cat((X, Y), dim=1)

X, Y, a, b
```

```python
# 逻辑运算符
X == Y
```

```python
# 对张量中的所有元素进行求和，会产生一个单元素张量
X.sum()
```

## 广播机制

即使形状不同，我们仍然可以通过调用*广播机制*（broadcasting mechanism）来执行按元素操作。


```python
X = torch.arange(3).reshape((3, 1))
Y = torch.arange(2).reshape((1, 2))

a = X+Y
X, Y, a
```

- (3,1),(1,2),都扩展为(3,2). 
- 矩阵`X`将复制列，变成`[[0,0], [1,1], [2,2]]`.
- 矩阵`Y`将复制行，变成`[[0,1], [0,1], [0,1]]`
- 然后相加

## 节省内存

下面将取消引用原来的张量，而是指向新分配的内存处的张量。

这可能是不可取的，原因有两个：首先，我们不想总是不必要地分配内存。
在机器学习中，我们可能有数百兆的参数，并且在一秒内多次更新所有参数。
通常情况下，我们希望原地执行这些更新。其次，如果我们不原地更新，其他引用仍然会指向旧的内存位置，这样我们的某些代码可能会无意中引用旧的参数。

```python
X = torch.randn(3,4)
Y = torch.randn(3,4)

# Y = X+Y, 取消Y
before1 = id(Y)
Y = Y + X
after1 = id(Y)

# Z = X+Y, 取消Z
Z = torch.zeros_like(Y)
before2 = id(Z)
Z = X + Y
after2 = id(Z)

before1 == after1, before2 == after2
```

`X[:] = X + Y`或`X += Y`来减少操作的内存开销。


```python
X = torch.randn(3,4)
Y = torch.randn(3,4)

# Y = X+Y, 取消Y
before1 = id(Y)
Y = Y + X
after1 = id(Y)

# Z = X+Y, 取消Z
Z = torch.zeros_like(Y)
before2 = id(Z)
Z = X + Y
after2 = id(Z)

before1 == after1, before2 == after2
```

## 转换为其他Python对象


将深度学习框架定义的张量[**转换为NumPy张量（`ndarray`）**]很容易，反之也同样容易。


```python
A = X.numpy()
B = torch.tensor(A)
A, B, type(A), type(B), id(A), id(B)
```

要(**将大小为1的张量转换为Python标量**)，我们可以调用`item`函数或Python的内置函数。


```python
a = torch.tensor([3.5])
a, a.item(), float(a), int(a)
```
