迭代器

```python
l1 = [11,22,33]

iter1 = iter(l1)

a1 = next(iter1)
b1 = next(iter1)

print(a1, b1)
```

```python
l2 = [11, 22, 33]

iter2 = iter(l2)

a2 = next(iter2)

# iter 和 for 循环不会干扰
for b2 in l2:
    print(a2, b2, a2 == b2)
    break
```
